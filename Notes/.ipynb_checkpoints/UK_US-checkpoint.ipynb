## Dataset 'UK_US'

Scott's spreadsheet "UK_US_Series_ALL.xlsx" contains GDP data for UK and US from <a href="https://measuringworth.com/ukgdp/result.php">MeasuringWorth.com</a>, all in 2010 USD. The UK 'consistent series' is from 1700 to the present, consistent in terms of the geographical area it is measuring. That is, it is for England, Wales, Scotland and Northern Ireland. Prior to 1801 Northern Ireland is assumed to be a constant share of UK GDP given a lack of available data.

Here are [UK_US dataset modeling](#modeling UK_US) and [UK_US dataset plots](#plotting UK_US). The results are exactly the same as from the previous data, except of course that the UK is now properly in 2010 USD.

[UK from 1801:](#UK 1801)
From its description, the truly clean part of the UK consistent series is only from 1801. Taking that as the UK data yields very successful modeling---the average deviation of model from data is only 9.31%, the largest deviation is 32.13%, in 1932 (Great Depression). Since 216 data points is plenty enough, this will be our standard UK dataset.

### GDP per Capita (GDPPC)
The spreadsheet provides also these data. [Plots of GDPPC data](#GDPPC data), especially using log scale for GDPCC, show a quantity that rises exponentially. So we can try modeling it with hindered growth. 

[GDP per Capita modeling](#GDPPC modeling) is reasonably successful. Both US and UK show unhindered exponential growth with constant annual growth rates of 1.71% and 1.14%, respectively.

### Populations
Get populations from the ratio GDP/GDPPC. From the UN site we have populations for 1950-2015. [Comparison with UN data](#UN) shows resonable agreement. Relative deviations are a fraction of a percent for UK. For the US, the deviations start at about 4% in 1950 and decrease to less than 1% by 1980. 

[Modeling Populations](#POP modeling) produces one of our best successes. All 227 points of the US population model are within 3% of the data, on average, the largest deviation is only 7.85% in 1885. What's more, the results are in perfect agreement with the UN population model that started only in 1950--the model predictions for 2050 are 408 vs 414 million. Both models state that currently the US population grows in the hindered mode, adding $r_uQ_h$ = 3.6 million (2.7 million for the UN dataset) per year. But the UN model could only tell us that hindering started long before 1950; it could not capture properly the individual parameters $r_u$ and $Q_h$ and gave them unreasonable values, such as $r_u$ = 11.69%. With the new extended dataset we now have $r_u$ = 3.2% and know that the hindering threshold $Q_h$ = 112 million was crossed in 1923. 

Hindered-growth modeling fails for the UK population so I tried [Linear model for the UK Population](#UK linear). This works well --- the average deviation of model from data is 3.63%, the largest deviation is 7.88% in 1914; from 1801 to 2016, population variation of the UK is well described as adding 256 thousand people every year.  

### Summary
The success of US population modeling, producing similar results from both large and limited datasets, in encouraging. In contrast, US GDP modeling with the limited dataset for 1967--2015 produced significant differences from the larger dataset. Presumably, the smooth population data yield more reliable model results than the noisy GDP data, which is not surprising.

Looking at the data alone, the [Growth Factors](#Growth Factors) show that GDP increases because both population and GDP/Capita are increasing. In the UK case, the GDPPC contribution to the growth was more than twice that of population.

Our models are internally self-consistent. GDPPC grows exponentially at a constant $r_u$ for both nations. Combined with linear population growth, the UK GDP grows unhindered, puerly exponentially. In the US, hindering of the population growth yields hindering of GDP. The primary quantities are, of course, population and GDP --- GDPPC is a derived quantity. Would be good if we could get directly the population data instead of deriving them in a roundabout way.

### Erratum

The last statement is wrong---it is impossible to get a self-consistent set of individual models for all three quantities. Denote GDP by $G$, population by $P$ and GDPPC by $C$, then $C = G/P$. Because the growth rate of quantity $Q$ is $r = d\ln Q/dt$, 

$$
    r_C = r_G - r_P = \frac{r_{G,u}}{1 + G/G_h} - \frac{r_{P,u}}{1 + P/P_h}   \qquad \qquad (1)
$$

where the last equality uses the 1st-order hindering parametrization of the growth rates for $G$ and $P$. Evidently, $r_C$ can have the same form if and only if the denominators of the two terms on the right are identical:

$$
   1 + G/G_h = 1 + P/P_h \qquad \Rightarrow \qquad C = \frac{G}{P} = \frac{G_h}{P_h}
$$

That is, $C$ is constant, $G$ is proportional to $P$. Since GDPPC is **not** constant, modeling it with the hindered growth parametrization is internally inconsistent. Such modeling was numerically possible only because of the noisy data, producing models with model-from-data average deviations of 16% and 24% for US and UK, respectively, and maximum deviations of 52% for the US and 47% for the UK.  

While reasonably acceptable numerically, the only self-consistent modeling of GDPPC is to mimic what's done with the data and take the [ratio of models for GDP and POP, as done below](#GDPPC model). This produces somewhat better fits, with deviations of model from data of 12% and 10% for US and UK, respectively, and largest deviations of 37% and 35%. An interesting result, seen in the [plots of GDP:POP model ratios](#GDPPC model plots), is that the growth rates of GDPPC show an **increase** with time, in nice agreement with the underlying 20-years rolling average of the data. Clearly, such an effect can never be captured by the hindering parametrization. In the US case, the growth rate of GDPPC rises continuously at a more or less constant rate, peaking at 2.3%/yr in 1945, slowly declining afterwards. In the case of the UK there's an initial rapid increase, followed by a more moderate rise with no clear sign of decrease, reaching a maximum of 1.7% in 2049. The difference in the behavior of GDPPC on one hand and of GDP and POP on the other is evident from the [plot of UK:US ratio](#UK:US comp) for all three quantities, both data and models.

# GDPPC model from GDP:POP model ratio <a id='GDPPC model'></a>

```python
#Get the model for GDPPC from model[GDP]/model[POP]:

print '\n***Modeling GDP per Capita from the ratio of models for GDP and Population'
PC_model, PC_rmodel = {}, {}
for nation in Nations:
    print '\n', nation, ':'
    PC_model[nation] = G_model[nation]/POP_model[nation]
    #Calculate the growth rate of the GDPPC/POP model:
    future_ = np.arange(Year[nation][0], 2051 )
    PC_rmodel[nation], _ = gr_data(PC_model[nation], future_ )   
    #and the modeling errors
    yf = np.searchsorted(future_, Year[nation][-1]) + 1
    err = 100*abs(GDPPC[nation]/PC_model[nation][:yf] - 1)
    
    print '***Average deviation of model from data is %4.2f'\
                %(np.average(err))+'%'
    print '   largest deviation of model from data is %4.2f'\
                %(np.amax(err))+'%' + ' in %i'%Year[nation][np.argmax(err)]
    print '***Growth rate reaches a peak of %4.2f'\
                %(np.amax(PC_rmodel[nation]))+'%/yr' + ' in %i'%future_[np.argmax(PC_rmodel[nation])]
```
